# 16- Splitting a string into multiple columns

```python
#import libraries
import pandas as pd 
import seaborn as sns

df = pd.DataFrame({'name': ['Adnan Ghumro', 'Farhan Ali', 'Perah Gh', 'Paras Gh'],
                   'location': ['Pisa Italy', 'Sukkur Pakistan', 'Karachi Pakistan', 'Hyderabad Pakistan']})
df
```

```python
df.name.str.split(' ').head()
```

```python
#splitting 1 col into two cols 
df.name.str.split(' ', expand = True)
```

```python
#adding splitted cols as new cols
df[["first_name", "last_name"]]=df.name.str.split(' ', expand = True)
df
```

# Assignment : do same for the location col. 

```python
# splitting location col into two cols
df.location.str.split(' ').head()
```

```python
#adding splitted cols as new cols (location)
df[["city_name", "country_name"]]=df.location.str.split(' ', expand = True)
df
```

```python
#refine data manipulation
df = df[['first_name', 'last_name', 'city_name', 'country_name']]
df
```
